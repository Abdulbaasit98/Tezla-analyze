```python
data = pd.read_csv("/content/drive/MyDrive/tesla.csv")
data
```

```python
data.set_index("Date", inplace= True)
```

# New Section

```python
data.index
```

```python
d_index=pd.to_datetime(data.index)
```

```python
data.set_index(d_index, inplace=True)
```

```python
data.index
```

```python
n_data=data.loc[data.index>"2016-01-01"]
n_data
```

```python
plt.figure(figsize=(8,6))
plt.plot(n_data["Open"])
plt.plot(n_data["Close"])
plt.title("Tesla aksiyalari: ochilishida")
plt.xlabel("Vaqt")
plt.ylabel("Narx")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(n_data["Open"], marker="o")
plt.plot(n_data["High"], marker="s")
plt.title("Tesla aksiyalari: ochilishida")
plt.xlabel("Vaqt")
plt.ylabel("Narx")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(n_data["Open"], linestyle = "-", c="blue", linewidth=1 )
plt.title("Tesla aksiyalari: ochilishida")
plt.xlabel("Vaqt")
plt.ylabel("Narx")
plt.show()
```

# New Section

```python
plt.figure(figsize=(8,6))
plt.plot(n_data["Open"], marker="s", linestyle = "-", c="blue", linewidth=1 )
plt.title("Tesla aksiyalari: ochilishida")
plt.xlabel("Vaqt")
plt.ylabel("Narx")
plt.grid("both")
plt.show()
```

```python
#bir necha grafikni bitta rasmga joylash
plt.figure(figsize=(6,4))
plt.subplot(1,2,1)
plt.plot(n_data["Open"], linestyle = "-", c="blue", linewidth=1 )
plt.title("Tesla aksiyalari: ochilishida")
plt.xlabel("Vaqt")
plt.ylabel("Narx")
plt.xticks(rotation=45)
plt.subplot(1,2,2)
plt.plot(n_data["Close"], linestyle = "-", c="red", linewidth=1 )
plt.title("Tesla aksiyalari: ochilishida")
plt.xlabel("Vaqt")
plt.ylabel("Narx")
plt.xticks(rotation=45)
plt.show
```
